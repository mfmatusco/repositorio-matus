```python
import pandas as pd
import re

archivo_entrada = 'hot100_since2000_to2025.csv'
archivo_salida = 'artistas_hot100.csv'

try:
    df = pd.read_csv(archivo_entrada, sep=',', on_bad_lines='skip')
except FileNotFoundError:
    print(f"Error: No se encontró el archivo '{archivo_entrada}'. Asegúrate de haberlo subido a la sesión de Colab.")
    exit()

collab_delimiters = r' Featuring | feat\. | Feat\. | With | And | & | \| '
artistas_individuales = []

for full_artist_name in df['Artist'].astype(str).unique():
    partes = re.split(collab_delimiters, full_artist_name, flags=re.IGNORECASE)

    for artista in partes:
        nombre_limpio = artista.strip()
        if nombre_limpio:
            artistas_individuales.append(nombre_limpio)

artistas_hot100 = sorted(list(set(artistas_individuales)))

df_artistas_individuales = pd.DataFrame(artistas_hot100, columns=['Artista Individual Unico'])

df_artistas_individuales.to_csv(archivo_salida, index=False)

print(f"¡Proceso completado! Se han identificado {len(artistas_hot100)} artistas individuales únicos.")
print(f"El listado final (separando colaboraciones) se ha guardado en el archivo: {archivo_salida}")
```
