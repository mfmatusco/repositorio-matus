```
import pandas as pd

nombre_archivo = 'hot100_since2000_to2025.csv'

try:
    df = pd.read_csv(nombre_archivo)
    print("Archivo cargado exitosamente.")
except FileNotFoundError:
    print(f"Error: No se encontró el archivo '{nombre_archivo}'. Asegúrate de subirlo.")

    exit()

df['Date'] = pd.to_datetime(df['Date'])

df_top10 = df[df['Rank'] <= 10].copy()

df_top10 = df_top10.sort_values(by=['Date', 'Rank'])

print("\nPrimeras 20 filas del Top 10 filtrado:")
print(df_top10.head(20))

conteo_por_fecha = df_top10.groupby('Date').size()
print("\nVerificación de canciones por fecha (deberían ser 10):")
print(conteo_por_fecha.head())

df_top10.to_csv('hot100_top10_only.csv', index=False)
print("\nArchivo guardado como 'hot100_top10_only.csv'")
```
